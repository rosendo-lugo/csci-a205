```python
# This program calculates miles per gallon.
# The results is displayed in an auto
# dialog box.

import tkinter

class MilesPerGallonCalculatorGUI:
    def __init__(self):
        # Create the main window.
        self.main_window = tkinter.Tk()
        
        # Set the title of the main window.
        self.main_window.title("Miles Per Gallon Calculator")
        
        # Create frames to group widgets.
        self.miles_frame = tkinter.Frame(self.main_window)
        self.gallon_frame = tkinter.Frame(self.main_window)
        self.calculate_frame = tkinter.Frame(self.main_window)
        self.result_frame = tkinter.Frame(self.main_window)
        self.bottom_frame = tkinter.Frame(self.main_window)
        
        # Widgets for the miles frame.
        self.miles_prompt_label = tkinter.Label(self.miles_frame, text='Enter the number of miles:')
        self.miles_entry = tkinter.Entry(self.miles_frame, width=10)
        self.miles_prompt_label.pack(side='left')
        self.miles_entry.pack(side='left')
        
        # Widgets for the gallon frame.
        self.gallons_prompt_label = tkinter.Label(self.gallon_frame, text='Enter the number of gallons:')
        self.gallon_entry = tkinter.Entry(self.gallon_frame, width=10)
        self.gallons_prompt_label.pack(side='left')
        self.gallon_entry.pack(side='left')
        
        # Calculate button.
        self.calc_button = tkinter.Button(self.calculate_frame, text='Calculate', command=self.calculate)
        self.calc_button.pack(side='left')
        
        # StringVar for output label.
        self.value = tkinter.StringVar()
        
        # Widgets for the results frame.
        # Widgets for the results frame.
        self.descr_prompt_label = tkinter.Label(self.result_frame, text='Result:')
        self.results_label = tkinter.Label(self.result_frame, textvariable=self.value, anchor='w', width=20)
        self.descr_prompt_label.pack(side='left')
        self.results_label.pack(side='left')
        
        # Quit button.
        self.quit_button = tkinter.Button(self.bottom_frame, text='Quit', command=self.main_window.destroy)
        self.quit_button.pack(side='left')
        
        # Pack the frames.
        self.miles_frame.pack()
        self.gallon_frame.pack()
        self.calculate_frame.pack()
        self.result_frame.pack()
        self.bottom_frame.pack()
        
        # Enter the tkinter main loop.
        tkinter.mainloop()

    def calculate(self):
        # Get values from entries.
        miles = float(self.miles_entry.get())
        gallons = float(self.gallon_entry.get())
        
        # Calculate miles per gallon.
        mpg = miles / gallons
        
        # Display result.
        self.value.set(mpg)

if __name__ == '__main__':
    mpg_calculator = MilesPerGallonCalculatorGUI()
```
